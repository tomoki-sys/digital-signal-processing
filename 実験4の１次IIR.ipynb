```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
T=1/8000
wc=1000
```

```python
def analize(Omega):

  Omega=Omega*1000*2*np.pi
  z=np.exp(1j*Omega*T)
  #Gz=1/((1+(2/T)*((1-z**(-1))/(1+z**(-1))))*(2/T)*np.tan(wc*T/2))
  Gz=1/((1+(2/T)*((1-z**(-1))/(1+z**(-1)))*(1/wc)))
  y=abs(Gz)

  return y
```

```python
x=np.arange(0.1,4.1,0.1)
y=np.vectorize(analize)(x)
print(y*0.5)
```

```python
plt.plot(x,y*0.5)
plt.grid()
plt.show()
```

```python
df=pd.DataFrame(y*0.5)
df.to_csv("実験4の１次IIR.csv",header=None,index=None)
```
